```python
import numpy as np
import random
from shapely.geometry import Point, MultiPoint
from scipy.optimize import differential_evolution as scipy_de
from shapely.ops import nearest_points

import voronoi_utils as vu
import solvers

def is_feasible(n, fixed, polygon, A_min, params, solver="custom"):
    if n == 0:
        _, areas = vu.get_voronoi_cells_and_areas(fixed, polygon)
        ok = min(areas) >= A_min
        print(f"[n=0] {'feasible' if ok else 'not feasible'}")
        if ok:
            vu.plot_voronoi(fixed, [], polygon)
        return ok, []

    minx, miny, maxx, maxy = polygon.bounds
    bounds = [(minx, maxx), (miny, maxy)] * n

    def loss_area_only(x_flat):
        pts = [(x_flat[i], x_flat[i+1]) for i in range(0, len(x_flat), 2)]
        _, areas = vu.get_voronoi_cells_and_areas(fixed + pts, polygon)
        return -min(areas)

    def loss_dist_only(x_flat):
        pts = [(x_flat[i], x_flat[i+1]) for i in range(0, len(x_flat), 2)]
        min_d = float("inf")
        # addedâ€“added
        for i in range(len(pts)):
            for j in range(i+1, len(pts)):
                d = np.linalg.norm(np.array(pts[i]) - np.array(pts[j]))
                min_d = min(min_d, d)
        # addedâ€“fixed
        for p in pts:
            for f in fixed:
                d = np.linalg.norm(np.array(p) - np.array(f))
                min_d = min(min_d, d)
        return min_d

    # combined objective: minimize negative(area + reg*dist)
    def loss_obj(x_flat):
        return loss_area_only(x_flat) - params["reg"] * loss_dist_only(x_flat)

    if solver == "custom":
        bx, bf = solvers.differential_evolution_custom(
            loss_obj, bounds, polygon, A_min, params
        )
    elif solver == "scipy":
        # early-exit callback checking both thresholds
        def _callback(xk, _=None):
            ma, md = solvers.compute_metrics(xk, params["fixed"], polygon)
            return (ma >= A_min) and (md >= params["l2_min"])

        res = scipy_de(
            loss_obj, bounds,
            maxiter=params["maxiter"],
            popsize=params["popsize"],
            seed=params.get("seed", None),
            updating="deferred",
            callback=_callback
        )
        bx, bf = res.x, res.fun
    elif solver == "qpso":
        bx, bf = solvers.qpso_solver(loss_obj, bounds, polygon, A_min, params)
    elif solver == "qpso2":
        bx, bf = solvers.qpso_pairwise_solver(loss_obj, bounds, polygon, A_min, params)
    elif solver == "ga":
        bx, bf = solvers.ga_solver(loss_obj, bounds, polygon, A_min, params)
    elif solver == "spsa":
        bx, bf = solvers.spsa_solver(loss_obj, bounds, polygon, A_min, params)
    else:
        raise ValueError(f"Unknown solver: {solver}")

    ma, md = solvers.compute_metrics(bx, params["fixed"], polygon)
    ok = (ma >= A_min) and (md >= params["l2_min"])
    print(f"[n={n}] {'feasible' if ok else 'not feasible'} "
          f"min_area={ma:.4f}, min_dist={md:.4f}")
    if ok:
        added = [(bx[i], bx[i+1]) for i in range(0, len(bx), 2)]
        vu.plot_voronoi(params["fixed"], added, polygon)
        return True, added
    return False, []

def find_max_additional(fixed, polygon, A_min, params, solver="custom"):
    print(f"Thresholds â†’ A_min={A_min:.4f}, l2_min={params['l2_min']:.4f}")
    ok, _ = is_feasible(0, fixed, polygon, A_min, params, solver)
    if not ok:
        return 0, []

    low, n = 0, 1
    best_added = []
    while True:
        ok, added = is_feasible(n, fixed, polygon, A_min, params, solver)
        if not ok:
            break
        low, best_added = n, added
        n *= 2

    print(f"Binary search between n={low} (feasible) and n={n} (infeasible)")
    lo, hi = low, n
    while lo < hi - 1:
        mid = (lo + hi) // 2
        ok, added = is_feasible(mid, fixed, polygon, A_min, params, solver)
        if ok:
            lo, best_added = mid, added
        else:
            hi = mid

    print(f"Result â†’ Max extra points: {lo}")
    return lo, best_added

# --------------------------
# Example usage
# --------------------------
if __name__ == "__main__":
    random.seed(30)
    np.random.seed(30)

    # Create random convex polygon
    pts = np.random.rand(30, 2)
    polygon = MultiPoint([tuple(p) for p in pts]).convex_hull

    # 5 fixed interior seeds
    fixed = []
    minx, miny, maxx, maxy = polygon.bounds
    while len(fixed) < 5:
        x, y = random.uniform(minx, maxx), random.uniform(miny, maxy)
        if polygon.contains(Point(x, y)):
            fixed.append((x, y))

    A_min = polygon.area / 15
    l2_min = min(maxx-minx, maxy-miny) / 10

    common = {"fixed": fixed, "l2_min": l2_min, "reg": 0.1, "seed": None}
    custom_params = {**common, "maxiter":200, "popsize":30, "F":0.9,  "CR":0.8}
    scipy_params  = {**common, "maxiter":200, "popsize":30}
    qpso_params   = {**common, "maxiter":200, "popsize":30, "alpha":0.85}
    ga_params     = {**common, "popsize":100, "ngen":80,
                     "cxpb":0.7, "mutpb":0.4, "sigma":0.05, "indpb":0.2}
    spsa_params   = {**common, "maxiter":150, "a":0.2, "c":0.2,
                     "alpha":0.602, "gamma":0.101, "restarts":5}

    param_map = {
        "spsa": spsa_params,
        "custom": custom_params,
        "scipy": scipy_params,
        "qpso": qpso_params,
        "qpso2": qpso_params,   # reuse qpso_params
        "ga": ga_params,
    }

    for solver, params in param_map.items():
        print(f"\n=== Solver: {solver} ===")
        find_max_additional(fixed, polygon, A_min, params, solver=solver)
```
