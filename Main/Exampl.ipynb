```python
import random
import numpy as np

from random_polygon import generate_random_convex_polygon_with_fixed
from losses import loss_type_1
from find_max_additional import find_max_additional
from plot_voronoi import plot_voronoi

# ─────────────────────────────────────────────────────────────────────────────
# 1) Generate a random convex polygon and fixed interior seeds
# ─────────────────────────────────────────────────────────────────────────────
POLY_SEED    = 43       # Seed for reproducibility
HULL_SAMPLES = 50       # Number of points to form the convex hull
NUM_FIXED    = 6        # Number of interior seed points

# Note: parameter names match the function signature
polygon, fixed = generate_random_convex_polygon_with_fixed(
    rand_seed=POLY_SEED,
    convexhull_param=HULL_SAMPLES,
    num_fixed=NUM_FIXED
)

# ─────────────────────────────────────────────────────────────────────────────
# 2) Define feasibility thresholds
# ─────────────────────────────────────────────────────────────────────────────
A_MIN  = 0.03   # Minimum required Voronoi cell area
L2_MIN = 0.10   # Minimum required pairwise distance

# ─────────────────────────────────────────────────────────────────────────────
# 3) Initial Voronoi plot (fixed seeds only)
# ─────────────────────────────────────────────────────────────────────────────
print(f"Initial Voronoi (fixed seeds only) | A_MIN = {A_MIN}, L2_MIN = {L2_MIN}")
# Use solver="—" to indicate this is pre-solver visualization
plot_voronoi(fixed, [], polygon, solver="—")

# ─────────────────────────────────────────────────────────────────────────────
# 4) Assemble the problem specification for the solvers
# ─────────────────────────────────────────────────────────────────────────────
problem = {
    "fixed":   fixed,
    "polygon": polygon,
    "A_min":   A_MIN,
    "l2_min":  L2_MIN,
    # Loss returns –(min_cell_area) or +inf if infeasible
    "loss":    lambda pts: loss_type_1(pts, polygon, L2_MIN)
}

# ─────────────────────────────────────────────────────────────────────────────
# 5) Configure solver parameters
# ─────────────────────────────────────────────────────────────────────────────
common_params = {
    "popsize": 40,
    "maxiter": 200,
    "alpha":   0.75,
    "seed":    123
}

ga_params = {
    **common_params,
    "ngen":  100,
    "sigma": 0.1,
    "indpb": 0.1,
    "cxpb":  0.5,
    "mutpb": 0.2
}

de_params = {
    "popsize": 15,   # DE population multiplier
    "maxiter": 200,
    "seed":    123
}

spsa_params = {
    "a":        0.1,
    "c":        0.1,
    "alpha":    0.602,
    "gamma":    0.101,
    "maxiter":  200,
    "restarts": 50,
    "seed":     123
}

# List of solvers and their parameter sets
solvers_to_run = [
    ("ga",            ga_params),
    ("de",            de_params),
    ("qpso",          common_params),
    ("qpso_pairwise", common_params),
    ("spsa",          spsa_params),
]

# ─────────────────────────────────────────────────────────────────────────────
# 6) Execute each solver and report results
# ─────────────────────────────────────────────────────────────────────────────
for solver_name, params in solvers_to_run:
    print(f"\n--- Solver: {solver_name} ---")
    config = {"name": solver_name, "params": params}

    # plot_each=True will display intermediate feasible configurations
    max_count, best_added = find_max_additional(
        problem,
        config,
        plot_each=True
    )

    print(f" → Max extra points: {max_count}")
```
